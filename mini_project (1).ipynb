```python
import pandas as pd
data=pd.read_csv("Student Depression Dataset.csv")
data
```

```python
data.isnull().any()
```

```python
data.drop(columns=["CGPA"],inplace=True)
data
```

```python
data.rename(columns={
    "Academic Pressure":"Academic_Pressure",
    "Study Satisfaction":"Study_Satisfaction",
    "Sleep Duration":"Sleep_Duration",
    "Dietary Habits":"Dietary_Habits",
    "Have you ever had suicidal thoughts ?":"Have_you_ever_had_suicidal_thoughts ?",
    "Work/Study Hours":"Study_Hours"

},inplace=True)
data
```

```python
data.drop(columns=["id"],inplace=True)
data
```

```python
data.drop(columns=["Job Satisfaction"],inplace=True)
data
```

```python
data.drop(columns=["Work Pressure"],inplace=True)
data
```

```python
data.drop(columns=["Financial Stress"],inplace=True)
data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
bins = [10, 20, 30, 40, 50, 60]
labels = ['10-20', '21-30', '31-40', '41-50', '51-60']
data['Age_Group'] = pd.cut(data['Age'], bins=bins, labels=labels)
sleep_by_age = data.groupby(['Age_Group', 'Sleep_Duration']).size().unstack(fill_value=0)
sleep_by_age.plot(kind='bar', figsize=(10,6))
plt.title("Sleep Duration across Age Groups")
plt.xlabel("Age Groups")
plt.ylabel("Number of Students")
plt.legend(title="Sleep Duration", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

interpretation

The bar chart compares the number of students across different age groups based on their average sleep duration. The sleep durations are categorized as Less than 5 hours, 5â€“6 hours, 7â€“8 hours, More than 8 hours, and Others.





```python
import pandas as pd
import matplotlib.pyplot as plt

sleep_by_degree_students = data.groupby(['Degree', 'Sleep_Duration']).size().unstack(fill_value=0)
sleep_by_degree_students.plot(kind='bar', figsize=(10, 6))
plt.title("Sleep Duration across Degree Students")
plt.xlabel("Degree")
plt.ylabel("Number of Students")
plt.legend(title="Sleep Duration", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Interpretation

The bar chart displays the sleep duration patterns of students from various academic degrees. The sleep durations are grouped into five categories â€” Less than 5 hours, 5â€“6 hours, 7â€“8 hours, More than 8 hours, and Others â€” while the x-axis lists different degree programs, ranging from undergraduate to postgraduate and doctoral levels.

```python
data["Dietary_Habits"].value_counts().plot(

      kind="pie",
      autopct="%1.1f%%",
      colors=["skyblue","lightgreen","orange","pink"])
plt.title("Dietary Habits of Students")
plt.ylabel("")
plt.show()
```

Interpretation

The pie chart presents the distribution of dietary habits among students, divided into four categories â€” Unhealthy, Moderate, Healthy, and Others.



```python
import pandas as pd
import matplotlib.pyplot as plt

suicidal_thoughts_of_degree_students = data.groupby(['Degree', 'Have_you_ever_had_suicidal_thoughts ?']).size().unstack(fill_value=0)
suicidal_thoughts_of_degree_students.plot(kind='bar', figsize=(10, 6))
plt.title("Suicidal thoughts of Degree Students")
plt.xlabel("Degree")
plt.ylabel("Number of Students")
plt.legend(title="Suicidal thoughts", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Interpretation

The chart compares the number of students across various degree programs who have and have not experienced suicidal thoughts. It provides insights into the prevalence of suicidal ideation among different academic streams.

```python
Suicidal_thoughts_by_gender = data.groupby('Gender')['Have_you_ever_had_suicidal_thoughts ?'].apply(lambda x: (x == 'Yes').sum())

plt.figure(figsize=(8, 8))
Suicidal_thoughts_by_gender.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'green'])
plt.title("Suicidal thoughts by Gender")
plt.ylabel("")
plt.show()
```

Interpretation

The pie chart illustrates the distribution of students who reported having suicidal thoughts, categorized by gender â€” Male and Female.

```python
import matplotlib.pyplot as plt

depression_by_degree = data.groupby('Degree')['Depression'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
depression_by_degree.plot(kind='bar', color='red')
plt.title("Total Depression Cases by Degree")
plt.xlabel("Degree")
plt.ylabel("Number of Depression Cases")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Interpretation: from this table we analyse total depression cases by degrees. it express more in class 12 and low in others.

```python
import pandas as pd
import matplotlib.pyplot as plt

Dietary_Habits_of_degree_students = data.groupby(['Degree', 'Dietary_Habits']).size().unstack(fill_value=0)

Dietary_Habits_of_degree_students.plot(kind='bar', figsize=(10,6))
plt.title("Dietary_Habits of Degree Students")
plt.xlabel("Degree")
plt.ylabel("Number of Students")

plt.show()
```

```python
import matplotlib.pyplot as plt

Depression=[1,2,3,4,5]
Age=['10-20', '21-30', '31-40', '41-50', '51-60']
plt.plot(Depression,Age,marker='o',color='red',linestyle='-')
plt.title("Line plot of Depression")
plt.xlabel("Depression")
plt.ylabel("Age")
plt.show()
```

Interpretation: The chart shows a line plot depicting the relationship between Age groups (on the y-axis) and Depression levels (on the x-axis). Each red point represents the average or relative depression level for a specific age group.
